# Multiple Variable Linear Regression in Tensorflow

```python
import tensorflow as tf
import numpy as np
import pandas as pd
rng = np.random
```

```python
#Loading the basketball dataset to a Pandas Dataframe
data = np.loadtxt("data.txt", dtype="float32", encoding="utf8")
print(data)
print("\nShape of the Dataset:", data.shape)
```

```python
X1_data = data[:,0]
X2_data = data[:,1]
Y_data = data[:,2]
n_samples = Y_data.shape
print(Y_data.shape)
X1_data.dtype
```

Using Stochastic Gradient Descent

```python
# Parameters
learning_rate = 0.001
training_epochs = 1000
display_step = 10

#tf Graph Input
X1 = tf.placeholder("float")
X2 = tf.placeholder("float")
Y = tf.placeholder("float")

# Set model weights
W1 = tf.Variable(rng.randn(), name="weight1")
W2 = tf.Variable(rng.randn(), name="weight2")
b = tf.Variable(rng.randn(), name="bias")

#Constructing a Linear Model with multiple attributes9
hypothesis = tf.add(tf.add(tf.multiply(W1, X1), tf.multiply(W2, X2)), b)

#Cost Function - Mean Squared Error
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Gradient descent
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
train = optimizer.minimize(cost)

#Initializing the variables
init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)

    # Fit all training data
    for epoch in range(training_epochs):
        for (x1, x2, y) in zip(X1_data, X2_data, Y_data):
            sess.run(train, feed_dict={X1: x1, X2: x2, Y: y})
            c = sess.run(cost, feed_dict={X1: X1_data, X2: X2_data, Y: Y_data})
            #Display logs per epoch step
            if (epoch+1) % display_step == 0:
                c = sess.run(cost, feed_dict={X1: X1_data, X2: X2_data,  Y: Y_data})
                print("Epoch:", '%04d' % (epoch+1), "cost=", "{:.9f}".format(c), "Weights=", sess.run(W1), sess.run(W2), "Bias=", sess.run(b))
      
    print ("Optimization Finished!")
    training_cost = sess.run(cost, feed_dict={X1: X1_data, X2: X2_data, Y: Y_data})
    print ("Training cost=", training_cost, '\n')
```
